```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
hr=pd.read_csv("HR_Data.csv")
hr.shape
```

```python
hr.head()
hr.columns
```

```python
print(hr.shape)
hr.isnull().sum()
hr.head()
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
le_Sales=LabelEncoder()
le_Salary=LabelEncoder()
hr["Sales_Occured_n"]=le_Sales.fit_transform(hr["Sales_Occured"])
hr["salary_n"]=le_Salary.fit_transform(hr["salary"])
```

```python
#One Hot Encoding
hr.head()
```

```python
x1=pd.get_dummies(hr["number_project"],drop_first=True)
x2=pd.get_dummies(hr["time_spend_company"],drop_first=True)
x3=pd.get_dummies(hr["Work_accident"],drop_first=True)
x4=pd.get_dummies(hr["promotion_last_5years"],drop_first=True)
x5=pd.get_dummies(hr["Sales_Occured"],drop_first=True)
x6=pd.get_dummies(hr["salary"],drop_first=True)
hr=pd.concat([hr,x1,x2,x3,x4,x5,x6],axis=1)
```

```python
hr.drop(["number_project","time_spend_company","Work_accident","promotion_last_5years","Sales_Occured","salary"],axis=1,inplace=True)
```

```python
hr.head()
#hr["Sales_Occured"].value_counts()
#hr["Sales_Occured_n"].value_counts()
#hr["salary"].value_counts()
hr["salary_n"].value_counts()
```

```python
#Removing Categorical Variables from the original Variables
hr.drop(["Sales_Occured","salary"],axis=1,inplace=True)
```

```python
#Normalizing the data
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
hr.head()
```

```python
hr["satis"]=scaling.fit_transform(hr[["satisfaction_level"]])
hr["last_eval"]=scaling.fit_transform(hr[["last_evaluation"]])
hr["monthly_sal"]=scaling.fit_transform(hr[["average_montly_hours"]])
hr.drop(["satisfaction_level","last_evaluation","average_montly_hours"],axis=1,inplace=True)
```

```python
x=hr.drop(["left"],axis=1)
y=hr["left"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(x_train,y_train)
```

```python
predictions=dtree.predict(x_test)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=250)
```

```python
rfc.fit(x_train,y_train)
```

```python
predictions=rfc.predict(x_test)
```

```python
#Importing Confusion matrix and classification report
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
#Checking the Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
#Creating Outlier Detection function
def detect_outlier(data):
    outlier=[]
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
```
